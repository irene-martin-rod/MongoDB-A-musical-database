<p> Musical genres count </p>

```python
#Complete name
list(db.users.aggregate([
    {"$project": {"Top 5 genres": 1}},
    {"$unwind": "$Top 5 genres"},
    {"$group": {"_id": "$Top 5 genres", "Count": {"$sum": 1}}},
    {"$sort": {"Count": -1}}
]))
```

```python
#Less specific
list(db.users.aggregate([
    {"$project": {
        "Top 5 genres": 1
    }},
    {"$unwind": "$Top 5 genres"},
    {
        "$group": {
            "_id": None,
            "Pop": {"$sum": {"$cond": [{"$regexMatch": {"input": "$Top 5 genres", "regex": "Pop", "options": "i"}}, 1, 0]}},
            #add regex to Match to look for coincidences with Pop, $cond to create a conditions: 1 if there is a coindicence and
            #0 if not. options i is to ignore upper and lower letters. Input specifized the enter string
            "Metal": {"$sum": {"$cond": [{"$regexMatch": {"input": "$Top 5 genres", "regex": "Metal", "options": "i"}}, 1, 0]}},
            "Rock": {"$sum": {"$cond": [{"$regexMatch": {"input": "$Top 5 genres", "regex": "Metal", "options": "i"}}, 1, 0]}},
        }
    }
]))
```

<p> Oldest and newest song

```python
#Oldest song
list(db.users.aggregate([
    {"$project": {
        "User": 1,
        "TopSongsArray": {"$objectToArray": "$Top 5 songs"} #project the Top 5 songs but using $objectToArray the
        #embebed documents are converted into an array to individualize each song
    }},
    {"$unwind": "$TopSongsArray"}, #undo the array in different documents
    {"$group": {
        "_id": "$_id",
        "User": {"$first": "$User"},
        "OldestSong": {"$min": "$TopSongsArray.v.Year"}
    }}
]))
```

```python
list(db.users.aggregate([
    {"$match": {"User": "JMMR"}},
    {"$project": {
        "_id": 0,
        "User": 1,
        "OldestSong": {
            "$objectToArray": {
                "$ifNull": ["$Top 5 songs", []] #ifNull is to manage cases with null values
            }
        }
    }},
    {"$unwind": "$OldestSong"},
    {"$match": {"OldestSong.v.Year": 1972}}, #do again a new match for select only the song with 1972 as a year
    {"$replaceRoot": {"newRoot": "$OldestSong.v"}} #replace the root document with all the information
]))
```

```python
#Newest songs
list(db.users.aggregate([
    {"$project": {
        "User": 1,
        "TopSongsArray": {"$objectToArray": "$Top 5 songs"}
    }},
    {"$unwind": "$TopSongsArray"},
    {"$group": {
        "_id": "$_id",
        "User": {"$first": "$User"},
        "NewestSong": {"$max": "$TopSongsArray.v.Year"}
    }}
]))
```

```python
list(db.users.aggregate([
    {"$match": {"User": "FCF"}},
    {"$project": {
        "_id": 0,
        "User": 1,
        "NewestSong": {
            "$objectToArray": {
                "$ifNull": ["$Top 5 songs", []]
            }
        }
    }},
    {"$unwind": "$NewestSong"},
    {"$match": {"NewestSong.v.Year": 2023}},
    {"$replaceRoot": {"newRoot": "$NewestSong.v"}}
]))
```
